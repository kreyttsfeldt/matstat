```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
norm_rv=sts.norm(loc=1,scale=2)
sample=norm_rv.rvs(size=100)
plt.hist(sample,bins=10)
```

```python
sum(sample)/100
```

```python
x=np.linspace(-6,8,100)#нарезание на 100 отрезков
pdf=norm_rv.pdf(x)
plt.plot(x,pdf)
plt.ylabel('$F(x)$')
plt.grid()#решетка
plt.xlabel('$x$')
```

```python
x=np.linspace(-6,8,100)
cdf=norm_rv.cdf(x)
plt.plot(x,cdf,label='теор норм распределение')
from statsmodels.distributions.empirical_distribution import ECDF
ecdf=ECDF(sample)
plt.step(ecdf.x,ecdf.y,label="ECDF")
plt.legend()
```

```python
kde=sts.gaussian_kde(sample)#сглаживание
plt.plot(x,kde(x),label="KDE")
pdf=norm_rv.pdf(x)
plt.plot(x,pdf,label='Теор-ая ф-я плотности')
plt.legend()
```
